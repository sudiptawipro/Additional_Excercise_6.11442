### 13

```python
def getyear(str):
    start=str.index('(')
    end=str.index(')')
    return str[start+1:end]
```

```python
moviesdf['year']=moviesdf.title.map(getyear)
```

### 14

#### Since I hace selected first 1000 records, I have to change the date (2000-02-01) to get the results

```python
import time as t

def tmconv(epoct):
    return t.strftime('%Y-%m-%d', t.localtime(epoct))

movie_rating['ts'] =movie_rating['timestamp'].map(tmconv)
movie_rating[movie_rating.ts > '2000-02-01']
```
